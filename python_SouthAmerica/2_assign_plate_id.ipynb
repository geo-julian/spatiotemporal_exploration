```python
deposits_file_name = 'input_data/SA_positives.csv'
#map_extent = [-180, -80, 0, 85], #North America
map_extent = [-85, -29, -55, 15] #south america

### Remove Mesozoic datapoints ###

SA_positives = pd.read_csv(deposits_file_name)

data = SA_positives[SA_positives['age'] < 67]
data = data.reset_index(drop=True)

### Continuing as Normal ###

#data=pd.read_csv(deposits_file_name)
lons=data['lon']
lats=data['lat']
ages=data['age']

#plot the map
fig = plt.figure(figsize=(16,12),dpi=150)
ax = plt.axes(projection=ccrs.PlateCarree())

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_left = False
gl.xlocator = mticker.FixedLocator([-180, -150, -120, -90, -60, -30, 0, 30, 60, 90, 120, 150, 180])
gl.ylocator = mticker.FixedLocator([-90, -75, -60, -45, -30, -15, 0, 15, 30, 45, 60, 75, 90])
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'color': 'gray', 'weight': 'bold'}
gl.ylabel_style = {'color': 'gray', 'weight': 'bold'}

ax.stock_img()
#ax.set_extent(map_extent)
ax.set_extent([-85, -29, -55, 15])
agegrid_cmap = Utils.get_age_grid_color_map_from_cpt('agegrid.cpt')
cb = ax.scatter(lons, lats, 50, marker='.',c=ages, 
                vmin=1, vmax=240, cmap=agegrid_cmap)
plt.title('Mineral Deposits South America - Ages')
fig.colorbar(cb, shrink=0.5, label='Age(Myr)')
plt.show()
```

```python
import pandas as pd
import pygplates
import Utils

SA_positives = pd.read_csv('input_data/SA_positives.csv', index_col='index')
SA_negatives = pd.read_csv('input_data/SA_negatives.csv')

SA_positives = SA_positives[SA_positives['age'] < 67]
SA_positives = SA_positives.reset_index(drop=True)

config_file_name = 'config_SA.json'


Utils.load_config(config_file_name)
Utils.print_parameters()


lons=SA_positives['lon']
lats=SA_positives['lat']
static_polygons = pygplates.FeatureCollection(Utils.get_parameter('static_polygons_file'))
rotation_model = pygplates.RotationModel(Utils.get_files(Utils.get_parameter('rotation_files')))
plate_ids = Utils.get_plate_id(lons.tolist(), 
                               lats.tolist(), 
                               static_polygons, 
                               rotation_model)


SA_positives['plate_id'] = plate_ids
SA_positives.to_csv('case_AREPS/coreg_input/SA/positive_deposits.csv', index=True, index_label='index')
print(' ')
print("Here are the positives (mineral deposits) from South America...")
display(SA_positives)

print(' ')
print("And now, the negatives (or volcanic rocks!) from South America...")


lons=SA_negatives['lon']
lats=SA_negatives['lat']
static_polygons = pygplates.FeatureCollection(Utils.get_parameter('static_polygons_file'))
rotation_model = pygplates.RotationModel(Utils.get_files(Utils.get_parameter('rotation_files')))
plate_ids = Utils.get_plate_id(lons.tolist(), 
                               lats.tolist(), 
                               static_polygons, 
                               rotation_model)


SA_negatives['plate_id'] = plate_ids
SA_negatives.to_csv('case_AREPS/coreg_input/SA/negative_deposits.csv', index=False)
#display(SA_negatives)
```
